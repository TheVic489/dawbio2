# M√†scares de selecci√≥

#### Mask = Objecte que amaga tota la informaci√≥ que no volem.

##### 


```python
# 1. Seleccionar Pokemons criteri num√®ric. HP>= 100
mask_hp_over100 = pokedex.loc[:, 'HP']>=100

display(pokedex.loc[mask_hp_over100, :])
```

```python
#3. Volem obtenir els pokemons que tinguin o b√© el tipus 1 o el tipus 2 Water.

mask5 = (pokedex.loc[:,"Type 1"] == "Water") | (pokedex.loc[:,"Type 2"] == "Water")
display(pokedex.loc[mask5,:])
```

```python
# 4. Volem els pokemons amb el hp igual 50, rempla√ßa els valors np.nan per no aplica.
pokedex2 = copy.deepcopy(pokedex)

mask_hp_less_50 = (pokedex2.loc[:,'HP'] < 50)

result = pokedex2.loc[mask_hp_less_50, ['HP']] = 50
display(pokedex2.loc[mask_hp_less_50, ['HP', 'Name'] ])
```

```python
# Fes que tots els pokemon de generaci√≥ 1 tinguin 20 punts m√©s d'HP.

pokedex2 = copy.deepcopy(pokedex)

# Creem m√†scara.
mask_1gen = (pokedex2.loc[ : , 'Generation' ] == 1)

# Pas 2. Substituim valors m√†scara.
pokedex2.loc[ mask_1gen, ['HP'] ] += 20 

# Pas 3. Validaci√≥. Consultem el resultat despr√©s d'aplicar la m√†scara.
display(pokedex2.loc[mask_1gen, : ])
```

#### Fes que tots els pokemon de tipus el√®ctric, que en vez de decirse electric, ahora
tendran un s√≠mbolo de un rayo.
üó≤ => U+1F5F2

```python
pokedex2 = copy.deepcopy(pokedex)

# M√†scara de substituci√≥ de valors.

# Pas 1. Aplicar la m√†scara per seleccionar els que s√≥n de tipus el√®ctric.
mask_electrics = (pokedex2.loc[ : , 'Type 1' ] == 'Electric') | (pokedex2.loc[ : , 'Type 2' ] == 'Electric')

# Pas 2. Aplicar la m√†scara; √©s a dir, modificar nom√©s les files del dataframe que compleixen la m√†scara.
#  df. üó≤ => U+1F5F2
pokedex2.loc[mask_electrics,['Type 1'] ] = "üó≤"
pokedex2.loc[mask_electrics,['Type 2'] ] = "üó≤"

# Pas 3. Validaci√≥. Consultem el resultat despr√©s d'aplicar la m√†scara.
print(pokedex2.loc[mask_electrics, : ])

mask9 = (pokedex2.loc[ : , 'Type 1' ] == 'Electric') | (pokedex2.loc[:, 'Type 2'] == 'Electric')

# df. üó≤ => U+1F5F2
pokedex2.loc[mask9,['Type 1'] ] = "üó≤"
pokedex2.loc[mask9, ["Type 2"]] = "üó≤"

display(pokedex2.loc[mask9, :])
```

### Fes que els valors np.nan que hi ha a la filera de Type 2 hi posi "Not Avaliable"


```python

pokedex3 = copy.deepcopy(pokedex)
# Primer mirem si hi ha han.
print("Nan values", pokedex['Type 2'].isnull().sum())

# Pas. Podem aplicar dir√®ctament la m√†scara, fillna el que fa √©s 
mask_replace_nan = pokedex3.loc[:, 'Type 2'].fillna('Not Aviable')

pokedex3.loc[:, ['Type 2']] = mask_replace_nan
# Pas 3. Validaci√≥. Consultem el resultat despr√©s d'aplicar la m√†scara.
display(pokedex3)
# mirar si √©s np.nan o no, i si ho hes ho canvia pel valor.
```
